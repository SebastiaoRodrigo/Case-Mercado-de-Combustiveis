*   Bibliotecas



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```



*   Importando os dados dos combustíveis.



```python
dados =  pd.ExcelFile('dados_desafiodatascientistintern_vendas_distribuidoras_anp.xlsx')
dados_gasolina = pd.read_excel(dados, sheet_name="gasolina")
dados_diesel = pd.read_excel(dados, sheet_name="diesel")
dados_etanol = pd.read_excel(dados, sheet_name="etanol")
```

```python
dados_diesel.rename(columns = {'2016':2016}, inplace = True)# Percebo que a coluna 2016, o formato está diferente das demais, com isso mudo para ficar igual.
dados_diesel.head()
```

```python
dados_gasolina.rename(columns = {'2016':2016}, inplace = True)
dados_gasolina.head()
```

```python
dados_etanol.rename(columns = {'2016':2016}, inplace = True)
dados_etanol.head()
```



*   Realizo, essa função abaixo, somente pegar os resultados de cada região ao longo desses 20 anos.



```python
def Anos(dados,string):
  valores = []
  for anos in range(2000,2021):
    for i in range(len(dados)):
      if dados['regiao'][i] == string:
        valores.append(dados[(anos)][i])
  return valores


```

```python
regiao = ['df','go','ma','mt','mg','pa','sp','to','br']
Anos_gas = []
Anos_dis = []
Anos_eta = []
for i in range(len(regiao)):
  Anos_gas.append(Anos(dados_gasolina,regiao[i]))
  Anos_dis.append(Anos(dados_diesel,regiao[i]))
  Anos_eta.append(Anos(dados_etanol,regiao[i]))
```



*   Plotando os gráficos dos resultados de cada região, com  isso conseguimos ter alguns resultados mais claros.



```python
fig, axs = plt.subplots(9, 1, figsize=(15, 25))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("", fontsize=18, y=0.92)
my_xticks = ['2000','2001','2002','2003','2004','2005','2006', "2007",'2008','2009','2010','2011','2012',"2013",'2014','2015','2016','2017','2018','2019','2020']
x1 = np.arange(0,252,12)
for i, rsi in enumerate(Anos_eta):
    axs[i].plot(rsi,label= "Etanol")
    axs[i].plot(Anos_dis[i],label="Diesel")
    axs[i].plot(Anos_gas[i],label = 'Gasolina')
    axs[i].set_title("Região: "+regiao[i],fontsize=10,fontweight='bold' )
    axs[i].set_xlabel('Anos') 
    axs[i].set_ylabel('Consumo') 
    axs[i].set_xticks(x1)
    axs[i].set_xticklabels(my_xticks)
    axs[i].legend()
plt.show()
```



*   Analisando os gráficos.




Podemos perceber simplesmente analisando os gráficos acima, que em todass as regiões do pais, conjuntamente com o brasil, o combustivel `Diesel`, apresenta uma sazonalidade, temos que em todo fim de ano as vendas apresentam uma queda, com isso podemos estimar que esse tempo de sazonalidade seja por volta de 12 meses.

Assim como a `Gasolina`, que apresenta uma sazonalidade com duração  de 12 meses nas regiões do Brasil, assim como no próprio `br`. 

O combustível `Etanol`, apresenta sazonalidade na região Brasil, podemos perceber que em cada inicio de ano, uma queda no consumo do mesmo acontece, com um tempo de doze meses.

*   Tamanho total do Mercado.



Pegando o consumo de todos os combustiveis ao longo dos anos e somando, teremos:

```python

total = Anos_gas + Anos_dis + Anos_eta
Dado_total = pd.DataFrame(total[8], columns=['Total'])
```

```python
Dado_total = Dado_total['Total'].to_frame()

Dado_total['Mov'] = Dado_total['Total'].rolling(12).mean()
```



*   O tamanho do mercado da soma dos três combustíveis relativo a cada ano



```python
def soma_meses(lista):#funcao para somar os valores de cada mes
  valor = []
  for i in range(0,len(lista),12):
    valor.append(sum(lista[i:i+12]))
  return valor
```

```python
Soma_meses = soma_meses(total[8])
plt.figure(figsize=(15, 7)) 
plt.bar(my_xticks, Soma_meses, color ='red',width = 0.4)
plt.title('Tamanho do mercado')
plt.show()
```

```python
Dado_total[['Total', 'Mov']].plot(label='RELIANCE',figsize=(16, 8))
plt.xticks(x1,my_xticks)
plt.xlim(-1,253)
plt.show()
```
